```python
import pandas as pd
```

```python
# make inconsistenct data
data = {
    'date': ['2022-01-01', '20-01-2022', '2022/01/03', '2022-21-04', '2022/21/05'],
    'country': ['USA', 'U.S.A', 'Unites states of america', 'America', 'Unites states'],
    'name': ['John', 'John', 'John', 'Joohn', 'Joohn'],
    'sale 2020': [100, 200, 300, None, 500],
    'sale 2021': [None, 200, 300, 400, 500],

}
# make pandas dataframe

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# Standardize date format
import pandas as pd

# make inconsistent data

# Standardize date format
df['date'] = pd.to_datetime(df['date'], errors='coerce').dt.strftime('%Y-%m-%d')

df.head()
```

```python
# impute NaN values of above columns with origional value with same formate
```

```python
# Harmonize name of country column
import pandas as pd

# Replace inconsistent values in the country column
df['country'] = df['country'].replace({
    'U.S.A': 'USA',
    'Unites states of america': 'USA',
    'America': 'USA',
    'Unites states': 'USA'
})

df.head()
```

```python
# correct typographical mistake
df['name']= df['name'].replace({'John': 'Jone Doe ', 'Joohn':'john Doe '})
df.head()
```

```python
# remove duplicate 
df = df.drop_duplicates()
df
```

```python
# remove duplicate 
df = df.drop_duplicates(subset='name')
df
```
