```python
# visualize the dataset using heatmap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Select the columns for heatmap
columns_for_heatmap = ['total_bill', 'tip', 'size', 'sex_encoded', 'encoded_smoker', 'encoded_day', 'encoded_time']

# Compute the correlation matrix
correlation_matrix = df[columns_for_heatmap].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of tip dataset \n After label encoded')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Assuming 'df1' is your DataFrame containing the Titanic dataset
# Define the transformer
transformer = ColumnTransformer(
    transformers=[
        ("OneHot", OneHotEncoder(), ['embarked'])
    ],
    remainder='passthrough'
)

# Apply the transformer to the DataFrame
transformed_data = transformer.fit_transform(df1)

# Extracting column names after transformation
encoded_column_names = transformer.named_transformers_['OneHot'].get_feature_names_out(['embarked'])

# Combining column names after transformation with remaining column names
all_column_names = list(encoded_column_names) + list(df1.columns.drop(['embarked']))

# Creating DataFrame with correct column names
df1 = pd.DataFrame(transformed_data, columns=all_column_names)
df_encoded
```


`get_dummies `in pandas is used for converting categorical variables into numerical format by creating dummy variables. It's helpful for preprocessing data before training machine learning models that can't handle categorical data directly. It's particularly useful for creating one-hot encoded features, improving model performance, and preserving the interpretability of individual categories
